```python
def file_read(filename):
    fp = open(filename,"r")
    data = fp.read()
    return data

def filter_content(fdd):
    num = []
    char = []
    spe_char = []
    for word in fdd:
        if word.isalnum():
            if word.isnumeric():
                num.append(word)
            else:
                char.append(word)
        else:
            spe_char.append(word)
    return num,char,spe_char

def main():
    fn = 'sample.txt'
    fd = file_read(fn)
    n,c,s = filter_content(fd)
    print("Numeric:","".join(n))
    print("Char:","".join(c))
    print("Special:","".join(s))

if __name__=='__main__':
    main()
```
